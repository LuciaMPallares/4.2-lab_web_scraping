#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
#your code
github2=req.get(url).text  #esto es siempre igual solo cambia la url

soup=bs(github2, 'html.parser')
soup
````

````python
name = soup.find_all('h1', class_='h3 lh-condensed') 
name_repo=[]

for n in name:
    name_repo.append(n.text.strip())
name_repo

````

````python
type(name_repo)
````

````python
import re
````

````python
patron = r"(?<= )\w+.\w+"  

nombres=[]                 #hay que hacer un for para iterar por cada elemento detro de la lista, no lo hace por lista entera

for i in name_repo:
    nombres.extend(re.findall(patron, i))  #uso extend porque final crea una lista de cada elemnto y append una lista
                                          #de cada lista. Extend en cambio quita la lista de cada elemento y crea un global
nombres
````
